# **DOCUMENTACION RUST**

# 1. Estructura y Filosofía
---
## Enfoque
Rust es un lenguaje de programación enfocado en la seguridad de memoria y el alto rendimiento. A diferencia de Go, Rust prioriza el control total sobre los recursos y la seguridad en tiempo de compilación.
- **Sin Garbage Collector:** Rust no utiliza un recolector de basura; en su lugar, emplea un control manual pero seguro de los recursos.
- **Sistema de Ownership:** utiliza un sistema de "propiedad" (ownership) que evita errores comunes de memoria en tiempo de ejecución.
### Hola Mundo
---
Utiliza la macro `println!` (notar el signo `!`) y, a diferencia de Go, requiere obligatoriamente el punto y coma (`;`) al final de la expresión.

```rust
fn main() { 
    println!("Hola MUNDO!"); 
}
```

# 2. Variables y Mutabilidad
La gestión de variables en Rust es más estricta para garantizar que el programador tenga control total sobre los datos.
- **Inmutabilidad por defecto:** En Rust, una vez que asignas un valor a una variable, no puedes cambiarlo a menos que lo indiques explícitamente.
- **Uso de mut:** Para permitir que una variable sea alterada, se debe usar la palabra clave `mut`.
- **Tipado e Inferencia:** Rust puede inferir tipos, pero permite anotaciones explícitas usando `:tipo`

```rust
let nombre = "María";      // Inmutable por defecto
let mut contador = 0;      // Mutable gracias a 'mut'
let edad: i32 = 25;        // Anotación de tipo explícita (i32)
```

# 3. Funciones y Retornos Implícitos
Rust introduce un concepto interesante en sus funciones: la última expresión puede devolverse sin usar la palabra clave `return`.

- **Firma de función:** Se usa fn para declarar la función y una flecha (`->`) para indicar el tipo de retorno.


- **Retorno Implícito:** Si la última línea de una función no tiene punto y coma, Rust la interpreta como el valor de retorno.


- **Advertencia:** Poner un punto y coma en esa última línea impide el retorno implícito y generaría un error si se espera un valor.

```rust
fn sumar(a: i32, b: i32) -> i32 { 
    a + b // Retorno implícito (sin ';' al final)
}
```

# 4. Estructuras condicionales
## `if` (Condicionales)
En Rust, el `if` no requiere paréntesis alrededor de la condición, pero las llaves `{}` son estrictamente obligatorias. Al ser una expresión, puedes usarlo para asignar valores a una variable, siempre que ambos bloques (`if` y `else`) retornen el mismo tipo de dato.

```rust
fn main() {
    let puntaje = 85;
    let tiene_bono = true;

    // Uso de 'if' como una expresión para asignar un valor
    let estado = if puntaje >= 90 {
        "Excelente"
    } else if puntaje >= 70 {
        "Aprobado"
    } else {
        "Reprobado"
    };

    println!("Resultado del estudiante: {}", estado);

    // Condicionales con lógica combinada
    if tiene_bono && puntaje < 100 {
        let puntaje_final = puntaje + 5;
        println!("Se aplicó bono. Puntaje final: {}", puntaje_final);
    } else {
        println!("No se aplicaron cambios al puntaje original.");
    }
}
```

## `loop` (Bucle Infinito)
El `loop` es un bucle que se ejecuta indefinidamente hasta que se encuentra una instrucción `break`. Es una característica única de Rust frente a Go (donde se usa un `for` vacío). Una de sus funciones más potentes es que el `break` puede devolver un valor hacia afuera del bucle.

```rust
// 1. Ciclo FOR: Itera de 1 a 5 (el 6 queda fuera)
prinfn main() {
    let mut contador = 0;
    let objetivo = 10;

    println!("Iniciando búsqueda de valor...");

    // El resultado de este bucle se guardará en 'resultado_final'
    let resultado_final = loop {
        contador += 1;
        println!("Iteración número: {}", contador);

        if contador == objetivo {
            // El 'break' detiene el bucle y envía el valor 'contador * 2'
            break contador * 2; 
        }
    };

    println!("El bucle terminó. El valor procesado es: {}", resultado_final);
}
```

## `while` (Bucle Condicional)
El `while` ejecuta el bloque de código mientras la condición sea verdadera. Es ideal cuando no conoces exactamente cuántas veces debe repetirse el ciclo, pero tienes una condición de parada clara.

```rust
fn main() {
    let mut combustible = 5;
    let mut distancia_recorrida = 0;

    println!("--- Simulador de Viaje ---");

    // Se ejecuta mientras la condición sea verdadera
    while combustible > 0 {
        distancia_recorrida += 50;
        combustible -= 1;

        println!("Kilómetros recorridos: {}. Combustible restante: {}L", 
                distancia_recorrida, combustible);

        if combustible == 1 {
            println!("¡Advertencia: Combustible bajo!");
        }
    }

    println!("El vehículo se ha detenido. Distancia total: {} km.", distancia_recorrida);
}
```
