# 1. Estructura y Filosofía
## Creado en Google
---
Go fue creado por Google con el objetivo de priorizar la simplicidad y la productividad en el desarrollo de software. Es un lenguaje especialmente recomendado para la creacion de microservicios, prototipos y servicios de sistema
### Hola Mundo
---
A diferencia de otros lenguajes, Go requiere de una estructura de paquetes clara y el uso de librerias estandar para la salida de datos.


````go
package main // Definicion del paquete principal
import "fmt" // Importacion de la libreria de formato

func main() {
	fmt.Println("HOLA mundo!")
}
````

#### *NOTA*
---
Go no requiere el uso de punto y coma (`;`) al final de las lineas para terminar las expresiones.

````go
// Funcion
func sum(a int, b int) int {
	return a + b // Retorno explícito obligatorio
}
fmt.Println(sum(5, 3)) // Llamada a la función

// Condicional
if lastName == "Butterfly" {
	fmt.Println("She's royalty... and magical")
} 

//Bucle for
for i := 0; i < 5; i++ {
	fmt.Println(i)
}
````

# 4. Variables y Tipado
## Declaración explicita
---
Se define el nombre, el tipo, y el valor.
## Declaración corta (Inferencia)
---
El compilador deduce el tipo automaticamente usando el operador `:=`.
## Mutabilidad
---
Las variables en Go son Mutables por defecto, lo que permite reasignar valores sin marcadores adicionales

````go
name := "Star" // Declaración corta con inferencia de tipo string

var age int = 16 // Declaracion explicita

lastName  := "Butterfly" // Declaración corta con inferencia de tipo string

fromMewni := true // Declaración corta con inferencia de boolean

age = 17 // Permitido: las variables son mutables por defecto

fmt.Println(name, lastName, age, fromMewni);
````
